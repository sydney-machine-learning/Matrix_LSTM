```python
import os
import numpy as np

data_path = '../datasets/North_Indian_Ocean/images'
files = os.listdir(data_path)

files = np.random.choice(files, 10)
```

```python
from PIL import Image

images = []

for file in files:
    path = os.path.join(data_path, file)
    arr = np.load(path)
    arr = arr.min(axis=0)
    # arr = arr[0]
    img = Image.fromarray(arr)
    images.append(img)

img
```

```python
import torch
import torchvision.transforms as T
to_tensor = T.ToTensor()
resize = T.Resize((64, 64))
to_pil = T.ToPILImage()
```

```python
for img in images[:5]:
    img.show()
```

```python
for img in images[:5]:
    to_pil(resize(to_tensor(img))).show()
```
